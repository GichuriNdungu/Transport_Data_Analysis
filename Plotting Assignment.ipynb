```python
max(fuel_econ.co2)
data = fuel_econ.co2
skewness = stats.skew(data)
print("Skewness:{}".format(skewness))
if skewness > 0:
    print('Right-skewed distribution')
elif skewness < 0 :
    print('left-skewed distribution')
else:
    print('Symetrical (no skew) distribution')
```

```python
# YOUR CODE HERE
binwidth = 60
plt.hist(fuel_econ.co2, edgecolor="black", bins=range(min(fuel_econ.co2), max(fuel_econ.co2) + binwidth, binwidth),color='green')
plt.xticks(range(30, max(fuel_econ.co2) + binwidth, binwidth))
plt.xlabel("CO2")
plt.ylabel("Frequency")
plt.show()
```

```python
# YOUR CODE HERE
plt.hist(fuel_econ.drive)
plt.xticks(rotation=60)
plt.ylabel("Frequency")
plt.xlabel("Drive")
plt.title('Distribution of vehicle Drive Types')
plt.show()
```
